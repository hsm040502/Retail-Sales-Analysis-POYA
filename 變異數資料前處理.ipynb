```python
import pandas as pd
data = pd.read_csv("final_result.csv")
data.head()
```

```python
import random
general_cities = data[data['地區'] == '一般縣市']['縣市'].unique()
random_cities = random.sample(list(general_cities), 6)
random_cities
```

```python
filtered_df = data[(data['地區'] == '直轄市') | (data['縣市'].isin(random_cities))]
filtered_df
```

```python
def sample_group(group):
    return group.sample(n=min(6, len(group)), random_state=10)
```

```python
data['組合'] = data['季度'] + "_" + data['地區'] + "_" + data['品項定位'] + "_" + data['款式']
data.head()
```

```python
result = data.groupby('組合', group_keys=False).apply(sample_group)
result.set_index("季度", inplace=True)
del result["組合"]
result
```
